# Fase 4 — Programación Orientada a Objetos

## Proyecto: Breast Cancer Wisconsin Diagnostic

En esta fase se implementan principios de Programación Orientada a Objetos (POO) sobre los componentes desarrollados en etapas anteriores.

Se incorporan mecanismos de:

- Encapsulamiento
- Herencia
- Polimorfismo
- Modularización

Lo anterior utilizando clases reutilizables para algoritmos de búsqueda y componentes del pipeline de procesamiento de datos.

```python
import sys
from pathlib import Path

# Agregar carpeta src al path
sys.path.append(str(Path.cwd().parents[1] / "src"))

from buscadores import BuscadorLineal, BuscadorBinario
from buscadores import Buscador
```

```python
#celda de prueba

datos = [10, 5, 3, 8, 20, 15]

lineal = BuscadorLineal(datos)
binario = BuscadorBinario(datos)

print("Resultado búsqueda lineal:", lineal.buscar(8))
print("Resultado búsqueda binaria:", binario.buscar(8))
```

```python
#polimorfismo

buscadores = [
    BuscadorLineal(datos),
    BuscadorBinario(datos)
]

for buscador in buscadores:
    resultado = buscador.buscar(8)
    print(
        f"{buscador.__class__.__name__}: "
        f"índice encontrado = {resultado}"
    )
```

## Principios POO implementados

### Encapsulamiento

Los datos se almacenan como atributos protegidos (`_datos`) dentro de cada objeto.

### Herencia

Las clases `BuscadorLineal` y `BuscadorBinario` heredan de la clase abstracta `Buscador`.

### Polimorfismo

Ambas clases implementan el método `buscar()`, permitiendo utilizar una interfaz común para distintos algoritmos.

### Reutilización

Los algoritmos desarrollados en la Fase 3 fueron encapsulados como clases reutilizables.

```python
#Caso normal

datos = [10, 5, 3, 8, 20, 15]

assert BuscadorLineal(datos).buscar(8) != -1
assert BuscadorBinario(datos).buscar(8) != -1

print("Caso normal OK")
```

```python
#caso límita

datos = []

assert BuscadorLineal(datos).buscar(8) == -1
assert BuscadorBinario(datos).buscar(8) == -1

print("Caso límite OK")
```

```python
#valor no existente

datos = [1, 2, 3, 4, 5]

assert BuscadorLineal(datos).buscar(999) == -1
assert BuscadorBinario(datos).buscar(999) == -1

print("Valor inexistente OK")
```

```python
#herencia

print(issubclass(BuscadorLineal, Buscador))
print(issubclass(BuscadorBinario, Buscador))
```

```python
#polimorfismo

buscadores = [
    BuscadorLineal(datos),
    BuscadorBinario(datos)
]

for buscador in buscadores:
    print(
        buscador.__class__.__name__,
        buscador.buscar(4)
    )
```
